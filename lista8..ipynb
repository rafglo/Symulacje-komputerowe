```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def empirical_char_func(sample,t):
    n = len(sample)
    return 1/n * np.sum(np.exp(t*1j*sample))
    
```

```python
t_k = 1000
t_0 = 0
n = 1000
lambda_pois = 2
ts = np.linspace(t_0,t_k,n)
emp1 = [empirical_char_func(np.random.poisson(lambda_pois, n), t) for t in ts]
def pois_char_func(t,lambda_):
    return np.exp(lambda_ * (np.exp(1j* t) - 1))
plt.plot(ts, emp1, label="empiryczna")
plt.plot(ts, pois_char_func(lambda_pois, ts), label="teoretyczna")
plt.legend(loc="best")
plt.title("Funkcja charakterytyczna rozk≈Çadu Poissona")
plt.show()
```
