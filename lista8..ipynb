```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def empirical_char_func(sample,t):
    n = len(sample)
    return 1/n * np.sum(np.exp(t*1j*sample))
    
```

```python
t_k = 1000
t_0 = 0
n = 1000
lambda_pois = 2
ts = np.linspace(t_0,t_k,n)
emp1 = [empirical_char_func(np.random.poisson(lambda_pois, n), t) for t in ts]
emp1_real, emp1_imag = np.real(emp1), np.imag(emp1)
def pois_char_func(t,lambda_):
    return np.exp(lambda_ * (np.exp(1j* t) - 1))
teo1 = pois_char_func(ts, lambda_pois)
teo1_real, teo1_imag = np.real(teo1), np.imag(teo1)
plt.scatter(emp1_real, emp1_imag, label="empiryczna")
plt.scatter(teo1_real, teo1_imag, label="teoretyczna")
plt.legend(loc="best")
plt.title("Funkcja charakterytyczna rozkadu Poissona")
plt.show()
```
